```python
import numpy as np
import matplotlib.pyplot as plt


y = np.linspace(-5,5,100)
x=(y*y)
plt.plot(x,y,label='$y^2=x$')
plt.xlabel('$x-axis$')
plt.ylabel('$y-axis$')

def line_gen(A,B):
  len =10
  dim = A.shape[0]
  x_AB = np.zeros((dim,len))
  lam_1 = np.linspace(0,1,len)
  for i in range(len):
    temp1 = A + lam_1[i]*(B-A)
    x_AB[:,i]= temp1.T
  return x_AB



#Plotting directrix
#equation of directrix is given x=1 & x=4
P=np.array([1,-1.4])
Q=np.array([4,-2.8])
A=line_gen(P,Q)
plt.plot(A[0,:],A[1,:],label='$directrix$')

plt.plot(1,-1.4,'o')
plt.text(1,-1.4,'P(1,-1.4)')
plt.plot(4,-2.8,'o')
plt.text(4,-2.8,'Q(4,-2.8)')

#plotting focus
plt.plot(-3,-9,'o',label='$focus$')
plt.text(-3,-9,'F(-3,-9)')
plt.legend(loc='best')
plt.grid()
plt.xlim([-4,5])
plt.ylim([4,-10])



plt.show()
```
