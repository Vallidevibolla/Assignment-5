```python
import numpy as np
import matplotlib.pyplot as plt


y = np.linspace(-5,5,100)
x=(y*y)
plt.plot(x,y,label='$y^2=x$')
plt.xlabel('$x-axis$')
plt.ylabel('$y-axis$')

def line_gen(A,B):
  len =10
  dim = A.shape[0]
  x_AB = np.zeros((dim,len))
  lam_1 = np.linspace(0,1,len)
  for i in range(len):
    temp1 = A + lam_1[i]*(B-A)
    x_AB[:,i]= temp1.T
  return x_AB

def parab_gen(y,a):
    x = y**2
    return x

#Plotting axis
R=np.array([1,0])
S=np.array([0,1])
RS=line_gen(R,S)
plt.plot(RS[0,:],RS[1,:],'--',label='(1  0)x=1')



A=np.array([4,0])
B=np.array([0,1])
AB=line_gen(A,B)
plt.plot(AB[0,:],AB[1,:],'--',label='(1  0)x=4')

#Plotting axis
K1=np.array([2,-2])
L1=np.array([4,-2])
K1L1=line_gen(K1,L1)
plt.plot(K1L1[0,:],K1L1[1,:],'--',label='K1, L1')

plt.plot(2,-2,'o',label='POI')
plt.text(2,-2,' K1 (2,-2)')
plt.plot(4,-2,'o',label='POI')
plt.text(4,-2,' L1(4,-2)')

#Plotting axis
K2=np.array([2,2])
L2=np.array([4,2])
K2L2=line_gen(K2,L2)
plt.plot(K2L2[0,:],K2L2[1,:],'--',label='K2, L2')

plt.plot(2,2,'o',label='POI')
plt.text(2,2,' K2 (2,2)')
plt.plot(4,2,'o',label='POI')
plt.text(4,2,' L2 (4,2)')

plt.plot(0,0,'o',color='r')
plt.text(-0.4,0.4,'c')


#plotting focus
plt.plot(2,0,'o',label='$focus$')
plt.text(2,0,'F(2,0)')
plt.plot(0,0,'o',label='vertex')
plt.text(0,0,' V(0,0)')

plt.legend(loc='best')
plt.grid()
plt.xlim([-6,10])
plt.ylim([4,-8])



plt.show()
```
